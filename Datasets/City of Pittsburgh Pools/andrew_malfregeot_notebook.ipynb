# Pittsburgh Pools
To find the best neighborhood in Pittsburgh, our group is looking at a variety of different metrics.
I took a look at data regarding pools in various Pittsburgh neighborhoods.
I think this is a good way of finding the best neighborhood because pools provide people a fun place to enjoy themselves and get good exercise.
To use this data to find the best neighborhood, I attempted to find which neighborhood had:
- The most pools
- The largest pool capacity (in gallons of water)


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "City Pools.csv"
pools_df = pd.read_csv(file_path)

#only looks at pools
pools_only = pools_df[pools_df["type"].str.lower() == "pool"]

#finds most pools
pool_counts = pools_only["neighborhood"].value_counts()
most_pools_neighborhood = pool_counts.idxmax()
most_pools_count = pool_counts.max()

#finds highest capacity
capacity_by_neighborhood = (pools_only.groupby("neighborhood")["capacity"].sum().sort_values(ascending=False))
highest_capacity = capacity_by_neighborhood.idxmax()
highest_capacity_value = capacity_by_neighborhood.max()

print("Neighborhood with the most pools: " + most_pools_neighborhood + " (" + str(most_pools_count) + ")")

print("Neighborhood with the highest total pool capacity: " + highest_capacity + " (" + str(highest_capacity_value) + ")")

plt.figure(figsize=(10,6))
pool_counts.sort_values(ascending=True).plot(kind='barh')
plt.title("Number of Pools by Neighborhood in Pittsburgh")
plt.xlabel("Number of Pools")
plt.ylabel("Neighborhood")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
capacity_by_neighborhood.sort_values(ascending=True).plot(kind='barh')
plt.title("Total Pool Capacity by Neighborhood in Pittsburgh")
plt.xlabel("Total Capacity (gallons)")
plt.ylabel("Neighborhood")
plt.tight_layout()
plt.show()
```

# Conclusion
With these metrics, I found that Highland Park is the best neighborhood when strictly looking at these pool-related measures. It tied South Side Flats for most pools in addition to having the largest capacity. Despite South Side Flats having the same number of pools, Highland Park almost doubles its capacity.
