```python
#Looking through the datasets, I believed that school enrollment was fairly important in considering a 
#'best' neighborhood.
import matplotlib.pyplot as plt
import pandas as pd
#This dataset is the data of the amount of enrolled students and total children in each respective neighborhood\
neighborhood = " "
max_ratio = -1
#I used the ratio of total enrolled students to the total amount of kids present within the neighborhood
enroll = pd.read_csv("school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over.csv")
for _, row in enroll.iterrows():
    ratio = row["Estimate; Enrolled in school:"] / row["Estimate; Total:"]
    if ratio > max_ratio:
        max_ratio = ratio
        neighborhood = row["Neighborhood"]
ratios = enroll["Estimate; Enrolled in school:"] / enroll["Estimate; Total:"]
plt.bar(enroll["Id"], ratios)
plt.xlabel("ID")
plt.ylabel("Enrollment Ratio")
plt.title("Enrollment Ratio by ID")
plt.show()
#In this graph, it visualizes the ratio by neighborhood, and each ID correlates to a neighborhood
#In the graph, we can see that id 20, which is Oakland, has the most amount of student enrolled compared to total children
print("The highest average of enrolled in school to total children is:", neighborhood, "with the average being:", max_ratio)
#While although data driven statistics can provide insight into which neighborhoods have over others, people can still have 
#subjective feelings about their favorite neighborhood. For example, I did not grow up in Pittsburgh, so I can more easily determine
#the 'best' neighborhood through statistics. However, the people who did grow up in Pittsburgh can grow attachment to where they
#grew up. 
```
